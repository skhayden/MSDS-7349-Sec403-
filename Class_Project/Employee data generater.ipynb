```python
import pandas as pd
import numpy as np
```

```python
#US Data set
from faker import Faker
fake = Faker('en_US')

#Initialize data frame
df = pd.DataFrame(columns=['First Name','Last Name','Date of Birth','Job','Address','Salary','SSN','Employee_Id', 'Office_Location','Building_number','Phone_Number','Email'])

#Regular workers
for i in range(20000):
    df = df.append({'First Name':fake.first_name(),'Last Name':fake.last_name(),'Date of Birth':fake.date_of_birth(),'Job':fake.job(),'Address': fake.address(),'Salary':np.random.randint(35000, 90000),'SSN': fake.ssn(),'Employee_Id':'P' + str(np.random.randint(10000, 99999)),'Office_Location':fake.city(),'Building_number':fake.building_number(),'Phone_Number':fake.phone_number(),'Email':fake.email()}, ignore_index=True)

#Managers    
for i in range(500):
    df = df.append({'First Name':fake.first_name(),'Last Name':fake.last_name(),'Date of Birth':fake.date_of_birth(),'Job':fake.job() + ' Sr.Manager','Address': fake.address(),'Salary':np.random.randint(100000, 200000),'SSN': fake.ssn(),'Employee_Id':'P' + str(np.random.randint(10000, 99999)),'Office_Location':fake.city(),'Building_number':fake.building_number(),'Phone_Number':fake.phone_number(),'Email':fake.email()}, ignore_index=True)

#Leadership    
for i in range(100):
    df = df.append({'First Name':fake.first_name(),'Last Name':fake.last_name(),'Date of Birth':fake.date_of_birth(),'Job':fake.job() + 'VP','Address': fake.address(),'Salary':np.random.randint(200000, 500000),'SSN': fake.ssn(),'Employee_Id':'P' + str(np.random.randint(10000, 99999)),'Office_Location':fake.city(),'Building_number':fake.building_number(),'Phone_Number':fake.phone_number(),'Email':fake.email()}, ignore_index=True)
        
    
    
    
    
```
